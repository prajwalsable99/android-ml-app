```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bmw.csv')
df.head()
```

```python
df=df.drop(['fuelType','tax'],axis='columns')

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_model=LabelEncoder()
df['model_type']=le_model.fit_transform(df['model'])
```

```python
df
```

```python
le_model_mapping=dict(zip(le_model.classes_,le_model.transform(le_model.classes_)))

le_model_mapping
```

```python
trans_dummy=pd.get_dummies(df['transmission'])
```

```python
trans_dummy
```

```python
df=pd.concat((df,trans_dummy),axis='columns')
```

```python
df
```

```python
df=df.drop(['model','transmission'],axis='columns')
df
```

```python
model_type=df.model_type
df=df.drop('model_type',axis='columns')
df.insert(0,'modeltype',model_type)
df
```

```python
inp=df.drop('price',axis='columns')
inp
```

```python
op=df.price
op
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inp,op,test_size=0.2,random_state=10)
```

```python
#print(f'x_train =\n{x_train}\ny_train\n{y_train}\nx_test=\n{x_test}\ny_test=\n{y_test} ')
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```
